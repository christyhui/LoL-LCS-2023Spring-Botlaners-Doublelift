```python
import pandas as pd
import numpy as np
```

```python
esports2022 = pd.read_csv("./2022_LoL_esports_match_data_from_OraclesElixir.csv")
```

```python
perry = esports2022[esports2022["playername"] == "Perry"]
```

```python
list(esports2022.columns)
```

## find the column that has 36 NAs

```python
perry.columns[list(perry.isna().sum()).index(36)]
```

## find the columns that have 67 NAs (instances of 67)

```python
indices = [ind for ind, ele in enumerate(perry.isna().sum()) if ele == 67]
```

```python
perry.columns[indices]
```

## delete columns that have these NAs (idc about 36 or 67)

```python
perry = perry.dropna(axis = 1)
perry
```

## delete columns that have opponent information in them

```python
perry.columns
```

### find the indices where the column name starts with "opp"

```python
[ind for ind, ele in enumerate(perry.columns) if "opp" in ele]
```

```python
perry.columns[[ind for ind, ele in enumerate(perry.columns) if "opp" in ele]]
```

### delete opponent columns

```python
perry.drop(perry.columns[[ind for ind, ele in enumerate(perry.columns) if "opp" in ele]], axis = 1, inplace = True)
```

```python
perry.columns
```

## tally champ appearance so we can use groupby.sum() to find the number of times the champ was picked

```python
perry["champ_appearances"] = [1]*67
```

```python
perry.groupby("champion").sum()
```

## some useful information SORTED BY CHAMP APPEARANCE (champ appearances, kills, deaths, assists)

```python
perry.groupby("champion").sum().sort_values("champ_appearances", ascending = False)[["champ_appearances", "kills", "deaths", "assists"]]
```

## all useful information when they're summed - NOT SORTED

```python
perry_champs_sum = perry.groupby("champion").sum()[["champ_appearances", 
                                                    "result",
                                                    "kills", 
                                                    "deaths", 
                                                    "assists",
                                                    "teamkills",
                                                    "teamdeaths",
                                                    'doublekills', 
                                                    'triplekills', 
                                                    'quadrakills', 
                                                    'pentakills',
                                                    "firstblood", 
                                                    "firstbloodvictim",
                                                    "barons"
                                                   ]].reset_index()
```

```python
perry_champs_sum.rename(columns = {"result":"wins"}, inplace = True)
```

```python
perry_champs_sum
```

## all useful information when theyre summed - SORTED BY CHAMP APPEARANCE

```python
perry_champs_sum.sort_values("champ_appearances", ascending = False).reset_index(drop = True)
```

## create win ratio column

```python
perry_champs_sum["win_ratio"] = perry_champs_sum["wins"]/perry_champs_sum["champ_appearances"]
perry_champs_sum
```

## all useful information when theyre summed - SORTED BY WIN RATIO

```python
perry_champs_sum.sort_values("win_ratio", ascending = False).reset_index(drop = True)
```

## create first blood rate column

```python
perry_champs_sum["fb_rate"] = perry_champs_sum["firstblood"]/perry_champs_sum["champ_appearances"]
perry_champs_sum
```

## all useful information when theyre summed - SORTED BY FIRST BLOOD RATE

```python
perry_champs_sum.sort_values("fb_rate", ascending = False).reset_index(drop = True)
```
