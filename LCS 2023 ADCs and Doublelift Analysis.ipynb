Notice how Doublelift is not in this list. That is because 2021 was the year he was retired (as was 2022). Let's look at the years in which he did play. Just so I don't have to write all of this again, let's create two functions to create a dataframe with the year he played and also the data frame specific to his games. Furthermore, I will add a game tally column to these data frames. This will come in handy later. This data frame will exclude opponent statistics.

```python
def create_league_pos_df(year, league, pos):
    year_df = pd.read_csv("./" + str(year) + "_LoL_esports_match_data_from_OraclesElixir.csv")
    league_df = year_df[year_df["league"] == league]
    pos_league_df = league_df[league_df["position"] == pos]
    pos_league_df["gametally"] = [1]*pos_league_df.shape[0]
    columns_to_drop = pos_league_df.columns[[ind for ind, ele in enumerate(pos_league_df.columns) if "opp" in ele]]
    return pos_league_df.drop(columns_to_drop, axis = 1)
```

```python
def create_player_df(year, player):
    year_df = pd.read_csv("./" + str(year) + "_LoL_esports_match_data_from_OraclesElixir.csv")
    player_df = year_df[year_df["playername"] == player]
    player_df["gametally"] = 1*player_df.shape[0]
    columns_to_drop = player_df.columns[[ind for ind, ele in enumerate(player_df.columns) if "opp" in ele]]
    return player_df.drop(columns_to_drop, axis = 1)
```

I am particularly interested in the last 3 years Doublelift played in (2018, 2019, and 2020). To keep things consistent to 2023, let's only look at Spring regular seasons. How does he place in the Spring regular season?

```python
def find_lcs_bot_data_regseason(df):
    lcs = df[(df["league"] == "NA LCS") | (df["league"] == "LCS")]
    lcs_bot = lcs[lcs["position"] == "bot"]
    lcs_bot_regseason = lcs_bot[lcs_bot["playoffs"] == 0]
    lcs_bot_regseason_spring = lcs_bot_regseason[lcs_bot_regseason["split"] == "Spring"]
    lcs_bot_regseason_spring["gametally"] = [1] * lcs_bot_regseason_spring.shape[0]
    return lcs_bot_regseason_spring
```

Let's group each year by the playername and sum them and average them by creating a function to do so.

```python
# returns two dataframes: one sum and one mean data frames
def create_sum_and_mean_df(df):
    # create sum df
    df_sum = df.groupby("playername").sum()[["gametally",
                                           "result",
                                           "kills",
                                           "deaths",
                                           "assists",
                                           "doublekills",
                                           "triplekills",
                                           "quadrakills",
                                           "pentakills",
                                           "firstblood",
                                           "firstbloodvictim",
                                           "totalgold",
                                           "goldat10",
                                           "xpat10",
                                           "golddiffat10",
                                           "xpdiffat10",
                                           "killsat10",
                                           "assistsat10",
                                           "deathsat10",
                                           "goldat15",
                                           "xpat15",
                                           "golddiffat15",
                                           "xpdiffat15",
                                           "killsat15",
                                           "assistsat15",
                                           "deathsat15"]]
    # create mean df
    df_mean = df.groupby("playername").mean()[["kills",
                                             "deaths",
                                             "assists",
                                             "firstblood",
                                             "firstbloodvictim",
                                             "dpm",
                                             "damageshare",
                                             "earned gpm",
                                             "cspm",
                                             "goldat10",
                                             "xpat10",
                                             "golddiffat10",
                                             "xpdiffat10",
                                             "killsat10",
                                             "assistsat10",
                                             "deathsat10",
                                             "goldat15",
                                             "xpat15",
                                             "golddiffat15",
                                             "xpdiffat15",
                                             "killsat15",
                                             "assistsat15",
                                             "deathsat15"]]
    
    # rename columns for sum
    for column in df_sum.columns:
        df_sum.rename(columns={column : str("total_") + column}, inplace = True)
        
    # add kda column for the sum df
    df_sum["kda"] = (df_sum["total_kills"] + df_sum["total_assists"])/df_sum["total_deaths"]
    
    # add winrate percent column for the sum df
    df_sum["winrate (%)"] = (df_sum["total_result"] / df_sum["total_gametally"]) * 100
        
    # rename columns for mean
    for column in df_mean.columns:
        df_mean.rename(columns={column : str("avg_") + column}, inplace = True)
        
    # add gametally column for the mean df
    df_mean["total_gametally"] = df.groupby("playername").sum()["gametally"]
    
    # add winrate percent column for the mean df
    df_mean["winrate (%)"] = (df.groupby("playername").sum()["result"]/df.groupby("playername").sum()["gametally"])*100
    
    return df_sum, df_mean
```

Let's create dictionaries for each of these 6 dataframes that sort by each column.

```python
def create_sorted_dict(df):
    sorted_dict = {}
    for column in df.columns:
        sorted_dict[column] = df.sort_values(column, ascending = False).reset_index()[["playername", "total_gametally", "winrate (%)", column]]
    # delete extra columns bc its useless, redundant, and will cause troubles when graphing
    del sorted_dict["total_gametally"]
    del sorted_dict["winrate (%)"]
    return sorted_dict
```

```python
sum_sorted_dict2018 = create_sorted_dict(lcsbot2018_sum)
sum_sorted_dict2019 = create_sorted_dict(lcsbot2019_sum)
sum_sorted_dict2020 = create_sorted_dict(lcsbot2020_sum)
mean_sorted_dict2018 = create_sorted_dict(lcsbot2018_mean)
mean_sorted_dict2019 = create_sorted_dict(lcsbot2019_mean)
mean_sorted_dict2020 = create_sorted_dict(lcsbot2020_mean)
```

Lets create graphs (and instead of showing them, simply print them into a folder to see) of each dictionary.

```python
# make sure you have the folder already named and created in the same directory as this .ipynb
def create_bar(sorted_dict, folder_name):
    for key in sorted_dict:
        plt.bar(sorted_dict[key]["playername"], sorted_dict[key][key])
        plt.xticks(sorted_dict[key]["playername"], rotation = "vertical")
        plt.ylabel(key)
        plt.xlabel("Player Name")
        plt.grid()
        # uncomment if you want to save the files into a folder
        plt.savefig("./" + folder_name + "/" + key + "_graph.png", dpi = 150, bbox_inches="tight", facecolor='white', transparent=False)
        plt.clf()
```
